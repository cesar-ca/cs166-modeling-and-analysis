Exercise 8: Random Walk with Drift

Use a biased coin to simulate a random walk of 30 steps on the line. If the coin falls head (H), take one step to the right, if it lands tails (T), take one step left. After 30 steps, note the final position. Take Pr(H) = 0.6 and Pr(T) = 0.4

(a) Plot a sample path.

(b) Make a histogram for 200 such random walks.

(c) Report the sample mean.

(d) Report the sample variance. (What should these be exactly?)

```python
# Python code for 1-D random walk.
import random
import numpy as np
import matplotlib.pyplot as plt
 
# Probability to move up or down
prob = [0.6, 0.4] 
```

```python
n = 30
p = 0.6
expected_value = n*(2*p - 1)
```

```python
position = 0
path = [(0,0)]
for i in range(30):
    generate_number = random.uniform(0,1)
    if generate_number <= 0.6:
        position += 1
    else:
        position -= 1

    path.append((i+1,position))
```

```python
def randomWalk():

    path = [(0,0)]
    position = 0
    for i in range(30):
        generate_number = random.uniform(0,1)
        
        if generate_number <= 0.6:
            position += 1

        else:
            position -= 1

        path.append((i+1, position))

    return path
```

PLOT A SAMPLE PATH

```python
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
x = [x[0] for x in path]
y = [x[1] for x in path]

plt.xlim(-4, 34)
plt.ylim(-4, 24)
plt.grid()
plt.plot(x, y, marker=".", markersize=20, markerfacecolor="green")
plt.show()
```

HISTOGRAM OF RANDOM WALKS

```python
d = []
for i in range(200):
    walks = randomWalk()[30][1]
    d.append(walks)
```

```python
n, bins, patches = plt.hist(x=d, bins = 12)
plt.xlim(-18, 28)
plt.ylim(0, 60)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Walks')
```

REPORT THE SAMPLE MEAN

```python
import statistics
mean_of_walks = statistics.mean(d)
print("The sample mean is ", mean_of_walks)
```

REPORT THE SAMPLE VARIANCE

```python
variance_of_walks = statistics.variance(d)
print("The sample variance is ", variance_of_walks)
```

Modify the simulation code in Section 1.1.3 starting on the page 8 for the gambler's ruin problem, but with various upper bounds on the iteration count. How does this bias the average duration of the game? How does this bias the sample variance?

```python
def randomWalkGambler():

    path = [(0,100)]
    position = 0
    for i in range(100):
        generate_number = random.uniform(0,1)
        
        if generate_number <= 0.6:
            position += 1

        else:
            position -= 1

        path.append((i+1, position))

    return path
```

```python
import random
position = 0
path_path = [(0,0)]
for i in range(20):
    generate_number = random.uniform(0,1)
    if generate_number <= 0.6:
        position += 1
    else:
        position -= 1

    path_path.append((i+1,position))
```

```python
def repeatFunction(upper_bound):
    position = 100
    path_path = [(0,100)]
    while position > 0 and position < upper_bound:
        generate_number = random.uniform(0, 1)
        if generate_number <= 0.5:
            position = position + 1
        else:
            position = position - 1
        path_path.append((path_path[-1][0] + 1, position))
    return path_path
```

```python
pathos_pathos = repeatFunction(2100)
```
